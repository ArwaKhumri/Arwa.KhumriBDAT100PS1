##Q3 Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 sides of a triangle and returns the area of the triangle. By Heron's formula, the area of a triangle with side lengths a, b, and c is s(s − a)(s − b)(s − c), where
s = (a + b + c)/2.

```
// Ans3)
public class Program
{
	public static void Main()
	{
		int a, b, c;
		Console.WriteLine("The the value of side 1");
		a = Convert.ToInt16(Console.ReadLine());
		Console.WriteLine("The the value of side 2");
		b = Convert.ToInt16(Console.ReadLine());
		Console.WriteLine("The the value of side 3");
		c = Convert.ToInt16(Console.ReadLine());
		double x = triangle(a, b, c);
		Console.Write(x);
	}
	public static double triangle(int a, int b, int c)
	{
		double s, area;
		s = (a + b + c) / 2;
		area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
		Console.Write(area);
		return area;
	}

}
```
